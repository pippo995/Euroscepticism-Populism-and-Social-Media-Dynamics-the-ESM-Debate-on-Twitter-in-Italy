```python
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
import re
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import cohen_kappa_score

pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_csv('TEST_sample.csv', low_memory=False)
```

```python
df
```

```python
mt = df[['cat','prev']]
confusion_matrix = pd.crosstab(mt['cat'], mt['prev'], rownames=['Actual'], colnames=['Predicted'])

ax= sn.heatmap(confusion_matrix, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 1, top - 1)
plt.show()
```

```python
C = confusion_matrix
```

```python
C
```

```python
D = C / C.astype(np.float).sum(axis=0)
```

```python
ax= sn.heatmap(D, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 1, top - 1)
plt.show()
```
